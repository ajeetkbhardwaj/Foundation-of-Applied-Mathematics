# Check primality
### A prime number 'n' has only 2 factors i.e. 1 and n
> The number 1 is NOT a prime number

```python
# to check wheather a number is prime or not
""" 
1. first define all the factors of number whom we needed to check primality
2. second check wheather number is prime or not?
3. third define the total prime upto that number
4. fourth define the total 
"""
def factors(n):
  factors = []

  for i in range(1, n + 1):
    if n % i == 0:
      factors.append(i)
  
  return factors
```

```python
def prime(n):
  return factors(n) == [1, n]
```

```python
print(prime(393))
factors(330)
```

```python
def primes_upto(m):
  primes = []

  for i in range(1, m + 1):
    if prime(i):
      primes.append(i)
  
  return primes
```

```python
def first_primes(m):
  count = 0
  i = 0
  primes = []

  while count < m:
    if prime(i):
      primes.append(i)
      count += 1
    i += 1
  
  return primes
```

```python
len(first_primes(2232))
sum(first_primes(2232))
```

```python
# A better implementation for the above code
def prime(n):
  flag = True
  i = 2

  while flag and i < n:
    if n % i == 0:
      flag = False
    i += 1
  
  return flag
```

```python
# An optimized implementation
import math

def prime(n):
  flag = True
  i = 2

  while flag and i < math.sqrt(n):
    if n % i == 0:
      flag = False
    i += 1
  
  return flag
```

```python
# Difference between two primes
def prime_diffs(n):
  previous = 2
  dict = {}

  for i in range(3, n + 1):
    if prime(i):
      diff = i - previous
      previous = i

      if diff in dict.keys():
        dict[diff] += 1
      else:
        dict[diff] = 1
  
  return dict
```
