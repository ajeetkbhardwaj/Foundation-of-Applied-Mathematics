## Compute GCD of two numbers

```python
# to define gcd of two integers

"""
1. Definition of function
2. list out all factors of both integers
3. list out all common factors in those factors
4. find the largest number from the list of common factor i.e gcd of both integers
"""
def gcd(m, n):
  common_factors = []

  for i in range(1, min(m, n) + 1):
    if m % i == 0 and n % i == 0:
      common_factors.append(i)
  
  return common_factors[-1]
```

```python
gcd(380,3424)
```

```python
# it is the improved version of the above program
""" 
1. all the procedures are same as previously except
2. storing the all factor and common factors in a list
3. we just take care of only largest common factor
"""
def gcd(m, n):
  for i in range(1, min(m, n) + 1):
    if m % i == 0 and n % i == 0:
      largest_common_factor = i
  
  return largest_common_factor
```

```python
print(f'GCD: {gcd(40, 60)}')
```
